Now we can create the $y$ function, plug in the limits, and compute the area under the "curve".  This is a stupid example that makes me laugh (motorcycles and math!) so it's a pretty trivial equation.

We know the answer is $b^2 - a^2$, so we'll plot it to understand why.

```python
from matplotlib.patches import Polygon

def func(x):
    return 2*x

def integrate(a,b,ax, title=''):
    x = np.arange(0,50,.1)
    y = func(x)
    ax.plot(x,y, lw=2) 
    ix = np.arange(a,b, .01)
    iy = func(ix)
    verts = [(a,0)] + list(zip(ix,iy)) + [(b,0)]
    poly = Polygon(verts, facecolor='.8')
    ax.yaxis.grid() 
    ax.xaxis.grid() 
    ax.add_patch(poly)
    
    ax.text(3,68,"$\int_{%d}^{%d}\!f(2x)\mathrm{d}x = %d$" 
            % (a,b, b**2 - a**2), fontsize=32)
    ax.set_title(title, fontsize=14)
    plt.suptitle('Integral of $\!f(2x)$',fontsize=18)
    
figsize(16,8)
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)

integrate(27, 28, ax1, 'with limits $27$ and $28$')
integrate(27-3, 28-2, ax2, 'with limits $27-3$ and $28-2$')
```
